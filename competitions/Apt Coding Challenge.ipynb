```python
def get_days(problems, k):
    count = 0
    new_problems = []
    while problems:
        i, j = 0, 1
        while j < len(problems):
            head, tail = problems[i], problems[j]
            if abs(head - tail) < k:
                new_problems.append(tail)
                j += 1
            else:
                i = j
                j += 1
        problems = list(new_problems)
        new_problems = []
        count += 1
    return count
```

```python
for _ in range(int(input())):
    n, k = map(int, input().split())
    vi = [int(i) for i in input().split()]

    m = [None] * n
    def match(s, v):
        if v[s]: return False
        v[s] = True
        for i in range(s+1, n):
            if abs(vi[i] - vi[s]) < k: continue
            if m[i] == None or match(m[i]):
                m[i] = s
                return True
        return False

    mc = 0
    for i in range(n):
        v = [False] * n
        if match(i, v): mc += 1
    print(n - mc)
```

```python
from itertools import combinations
def check(problems, n, k):
    for sub_set in combinations(problems, n):
        valid = True
        for i in range(len(sub_set) - 1):
            if abs(sub_set[i] - sub_set[i+1]) < k:
                valid = False
        if valid:
            return sub_set
    return False
def get_days(problems, k):
    count = 0
    n = len(problems)
    while problems:
        rs = check(problems, n, k)
        if rs:
            count += 1
            problems = set(problems) - set(rs)
            n = len(problems)
        else:
            n -= 1
    return count
```

```python
import math
value = 8
def get_sq(n, m):
    if n < m:
        n, m = m, n
    return m * (m+1) * (2*m+1)/6 + (n-m) * m * (m+1)/2 

def get_n(m, x):
    return (x * 6 + (m ** 3 - m)) / ((3 * m ** 2) + 3 * m)

def sqs(n, value):
    if get_sq(n, 2) > value:
        return [], True
    start, end = 2, value//2
    while end >= start:
        mid = (start + end) // 2
        val = get_sq(n, mid)
        if val == value:
            return (n, mid), False
        elif val > value:
            end = mid - 1
        else:
            start = mid + 1
    return [], False

result = []
seen = set()
for n in range(int(math.sqrt(value)) + 1):
    rs = sqs(n, value)
    if rs[0]:
        n, m = rs[0]
        if n in seen:
            break
        
        result.append((n, m))
        if n != m:
            result.append((m, n))
        seen.add(n)
        seen.add(m)
    if rs[1]:
        break

if value > 1 and (1, value) not in result:
    result = result + [(1, value), (value, 1)]

result.sort()
print(len(result))
for item in result:
    print(*item)
```
